Import Bible data from https://raw.githubusercontent.com/mxw/grmr/master/src/finaltests/bible.txt and do text preprocessing step by step

```python
import requests

url = "https://raw.githubusercontent.com/mxw/grmr/master/src/finaltests/bible.txt"
response = requests.get(url)
bible_text = response.text

print(f"Successfully downloaded data. First 500 characters:\n{bible_text[:500]}...")
print(f"Total length of text: {len(bible_text)} characters")
```

```python
import re
import requests # Ensure requests is available if needed

# Check if bible_text is defined, if not, re-fetch it
# This makes the cell robust if the previous cell wasn't run.
try:
    _ = bible_text
except NameError:
    print("Warning: 'bible_text' not found. Re-fetching data.")
    url = "https://raw.githubusercontent.com/mxw/grmr/master/src/finaltests/bible.txt"
    response = requests.get(url)
    bible_text = response.text

# Convert to lowercase
normalized_text = bible_text.lower()

# Remove non-alphanumeric characters (keeping spaces)
normalized_text = re.sub(r'[^a-z0-9\s]', '', normalized_text)

print(f"Normalized text (first 500 characters):\n{normalized_text[:500]}...")
print(f"Total length of normalized text: {len(normalized_text)} characters")
```

```python
from nltk.tokenize import word_tokenize
import nltk

# Download the 'punkt_tab' tokenizer data if not already downloaded
# This resource is specifically required by the PunktTokenizer used by word_tokenize.
try:
    nltk.data.find('tokenizers/punkt_tab/english/') # Check for the specific resource path
except LookupError:
    print("Downloading 'punkt_tab' resource for NLTK word_tokenize...")
    nltk.download('punkt_tab') # Download the specific resource

# Tokenize the normalized text
tokens = word_tokenize(normalized_text)

print(f"First 20 tokens: {tokens[:20]}")
print(f"Total number of tokens: {len(tokens)}")
```

```python
from nltk.corpus import stopwords
import nltk

# Download the 'stopwords' corpus if not already downloaded
try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    print("Downloading 'stopwords' corpus for NLTK...")
    nltk.download('stopwords')

# Get English stopwords
stop_words = set(stopwords.words('english'))

# Remove stopwords from the tokens
filtered_tokens = [word for word in tokens if word not in stop_words]

print(f"First 20 filtered tokens: {filtered_tokens[:20]}")
print(f"Total number of filtered tokens: {len(filtered_tokens)}")
print(f"Reduction in tokens: {len(tokens) - len(filtered_tokens)}")
```

```python
from nltk.stem import PorterStemmer

# Initialize the Porter Stemmer
stemmer = PorterStemmer()

# Apply stemming to the filtered tokens
stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]

print(f"First 20 stemmed tokens: {stemmed_tokens[:20]}")
print(f"Total number of stemmed tokens: {len(stemmed_tokens)}")
```

```python
from nltk.stem import WordNetLemmatizer
import nltk

# Download the 'wordnet' corpus if not already downloaded
try:
    nltk.data.find('corpora/wordnet')
except LookupError:
    print("Downloading 'wordnet' corpus for NLTK...")
    nltk.download('wordnet')

# Initialize the WordNet Lemmatizer
lemmatizer = WordNetLemmatizer()

# Apply lemmatization to the filtered tokens (or stemmed tokens, depending on desired order)
# It's generally better to lemmatize after stopword removal but before or after stemming depends on the use case.
# For this step, I'll lemmatize the filtered_tokens directly.
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

```

```python
print(f"First 20 lemmatized tokens: {lemmatized_tokens[:20]}")
print(f"Total number of lemmatized tokens: {len(lemmatized_tokens)}")
```
