```python
import zipfile
import os

zip_file_path = 'opinions.zip'
extract_dir = 'opinions_unzipped'

# Create the directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print(f"'{zip_file_path}' unzipped to '{extract_dir}'.")

# List the contents of the unzipped directory
print("Contents of the unzipped directory:")
for root, dirs, files in os.walk(extract_dir):
    for name in files:
        if name.endswith('.pdf'):
            print(os.path.join(root, name))
```

```python
import fitz # PyMuPDF

pdf_path = 'opinions_unzipped/13-1314_3ea4.pdf'

try:
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    print(f"Extracted text from {pdf_path}. First 500 characters:\n{text[:500]}...")
    # Store the extracted text in a variable for further processing
    extracted_pdf_text = text
except FileNotFoundError:
    print(f"Error: The file {pdf_path} was not found. Please ensure the path is correct.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import re
import string

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Remove numbers (optional, depending on the task)
    text = re.sub(r'\d+', '', text)
    # Remove extra whitespaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text

if 'extracted_pdf_text' in locals():
    preprocessed_text = preprocess_text(extracted_pdf_text)
    print(f"Original text length: {len(extracted_pdf_text)}")
    print(f"Preprocessed text length: {len(preprocessed_text)}")
    print(f"First 500 characters of preprocessed text:\n{preprocessed_text[:500]}...")
else:
    print("Error: 'extracted_pdf_text' not found. Please ensure the previous cell was executed successfully.")
```

```python
from nltk.tokenize import word_tokenize
import nltk

# Download the 'punkt' tokenizer if not already downloaded
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')

# Download the 'punkt_tab' resource as it's often implicitly required by word_tokenize
try:
    nltk.data.find('tokenizers/punkt_tab')
except LookupError:
    nltk.download('punkt_tab')

if 'preprocessed_text' in locals():
    tokens = word_tokenize(preprocessed_text)
    print(f"Number of tokens: {len(tokens)}")
    print(f"First 20 tokens: {tokens[:20]}")
else:
    print("Error: 'preprocessed_text' not found. Please ensure the previous cell was executed successfully.")
```

```python
from nltk.corpus import stopwords
import nltk

# Download the 'stopwords' corpus if not already downloaded
try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')

if 'tokens' in locals():
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]

    print(f"Number of tokens before stop word removal: {len(tokens)}")
    print(f"Number of tokens after stop word removal: {len(filtered_tokens)}")
    print(f"First 20 filtered tokens: {filtered_tokens[:20]}")
```

```python
from nltk.stem import WordNetLemmatizer
import nltk

# Download the 'wordnet' corpus for lemmatization if not already downloaded
try:
    nltk.data.find('corpora/wordnet')
except LookupError:
    nltk.download('wordnet')

# Download 'omw-1.4' for WordNet if not already downloaded
try:
    nltk.data.find('corpora/omw-1.4')
except LookupError:
    nltk.download('omw-1.4')


if 'filtered_tokens' in locals():
    lemmatizer = WordNetLemmatizer()
    lemmas = [lemmatizer.lemmatize(token) for token in filtered_tokens]

    print(f"Number of tokens after lemmatization: {len(lemmas)}")
    print(f"First 20 lemmas: {lemmas[:20]}")
else:
    print("Error: 'filtered_tokens' not found. Please ensure the previous cell was executed successfully.")
```

```python
from collections import Counter

if 'lemmas' in locals():
    word_counts = Counter(lemmas)

    print(f"Total unique lemmas: {len(word_counts)}")
    print("\nMost common 20 lemmas:")
    for word, count in word_counts.most_common(20):
        print(f"'{word}': {count}")
else:
    print("Error: 'lemmas' not found. Please ensure the previous cell was executed successfully.")
```
