```python
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv("heart_disease_uci.csv")
df2 = df.dropna(subset=["chol", "trestbps", "age", "thalch", "num"])
mean = np.mean(df2["chol"])
median = np.median(df2["chol"])
var = np.var(df2["chol"])
q1 = np.percentile(df2["chol"], 25)
q3 = np.percentile(df2["chol"], 75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
outliers = df2[(df2["chol"] < lower) | (df2["chol"] > upper)]
binom = stats.binom.pmf(5, 20, df2["num"].mean())
poisson = stats.poisson.pmf(2, 3)
z = stats.zscore(df2["trestbps"])
corr = np.corrcoef(df2["age"], df2["thalch"])[0,1]
mean_chol = np.mean(df2["chol"])
sem = stats.sem(df2["chol"])
ci = stats.norm.interval(0.95, loc=mean_chol, scale=sem)
ttest = stats.ttest_1samp(df2["trestbps"], 120)
table = pd.crosstab(df2["sex"], df2["num"])
chi2 = stats.chi2_contingency(table)
groups = [df2[df2["sex"] == g]["chol"] for g in df2["sex"].unique()]
fstat = stats.f_oneway(*groups)
print("\n\nMean, Median and variance:\n", mean, median, var)
print("\n\nOutliers:\n",outliers)
print("\n\nBinomial:\n",binom)
print("\n\nPoissons:\n",poisson)
print("\n\nZ-TEST:\n",z)
print("\n\nCorrelation:\n",corr)
print("\n\nConfidence Interval:\n",ci)
print("\n\nT-TEST:\n",ttest)
print("\n\nChi2 TEST:\n",chi2)
print("\n\nANOVA1 TESt:\n",fstat)
```
