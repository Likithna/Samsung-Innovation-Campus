## Requried Packages

```
import numpy as np
import matplotlib.pyplot as plt

# Data which will go through activations
x = np.linspace(-10,10,100)
```

## LeakyRelu Activation Function Formula

$f(x)= $
$\begin{cases}
\begin{align}
    z  \ &: \ z > 0\\
    \ alpa * z \ &: \ z \leq 0\\
    \end{align}
\end{cases}$

```
alpha = 0.01

def leakyrelu(x, Derivative=False):
  if not Derivative:
    return max(alpha * x, x)
  else:
    return 1 if x > 0 else alpha
```

## LeakyRelu Activation Function Graph

```
plt.figure(figsize=(12,8))
plt.plot(x, list(map(lambda x: leakyrelu(x),x)), label="leakyrelu")
plt.plot(x, list(map(lambda x: leakyrelu(x, True),x)), label="derivative")
plt.title("Leakyrelu Function (dataaspirant.com)")
plt.legend()
plt.show()
```
