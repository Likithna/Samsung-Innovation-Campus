## Requried Packages

```
import numpy as np
import matplotlib.pyplot as plt

# Data which will go through activations
x = np.linspace(-10,10,100)
```

## Linear Activation Function

$f \left( x_{i} \right) = x_{i}$

```
def linear(x, Derivative=False):
    if not Derivative:
        return x
    else:
        return 1.0
```

```
plt.figure(figsize=(12,8))
plt.plot(x, list(map(lambda x: linear(x),x)), label="linear")
plt.plot(x, list(map(lambda x: linear(x, True),x)), label="derivative")
plt.title("Linear Activation Graph (dataaspirant.com)")
plt.legend()
plt.show()
```
