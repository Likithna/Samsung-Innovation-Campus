## Requried Packages

```
import numpy as np
import matplotlib.pyplot as plt

# Data which will go through activations
x = np.linspace(-10,10,100)
```

```
x
```

## Relu Activation Function Formula

$f(x) = max(x, 0)$

```
def relu(x, Derivative=False):
    if not Derivative:
        return np.maximum(0,x)
    else:
        out = np.ones(x.shape)
        out[(x < 0)]=0
        return out
```

```
relu(x)
```

## Relu Activation Function Graph

```
plt.figure(figsize=(12,8))
plt.plot(x, list(map(lambda x: relu(x),x)), label="relu")
plt.plot(x, list(map(lambda x: relu(x, True),x)), label="derivative")
plt.title("Relu Function Graph (dataaspirant.com)")
plt.legend()
plt.show()
```
