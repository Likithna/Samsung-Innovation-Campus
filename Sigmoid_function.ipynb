## Requried Packages

```
import numpy as np
import matplotlib.pyplot as plt

# Data which will go through activations
x = np.linspace(-10,10,100)
```

## Sigmod Function Formula

$f\left(x_{i} \right) = \frac{1}{1 + e^{  - x_{i}  }}, \ \
f^{\prime}\left( x_{i} \right) = \sigma(x_{i}) \left( 1 -  \sigma(x_{i}) \right)$

```
def sigmoid(x,Derivative=False):
    if not Derivative:
        return 1 / (1 + np.exp (-x))
    else:
        out = sigmoid(x)
        return out * (1 - out)
```

## Sigmod Function Graph

```
plt.figure(figsize=(12,8))
plt.plot(x, list(map(lambda x: sigmoid(x),x)), label="sigmoid")
plt.plot(x, list(map(lambda x: sigmoid(x, True),x)), label="derivative")
plt.title("Sigmoid Function (dataaspirant.com)")
plt.legend()
plt.show()
```
