```
import numpy as np
import matplotlib.pyplot as plt

# Data which will go through activations
x = np.linspace(-10,10,100)
```

## Step Function Formula

$f(x)=$
\begin{cases}
\begin{align}
    0  \ &: \ x_{i} \leq T\\
    1 \ &: \ x_{i} > T\\
    \end{align}
\end{cases}

```
def step(x, threshold):
  return 0 if x < threshold else 1
```

```
threshold = 0
plt.figure(figsize=(12,8))
plt.plot(x, list(map(lambda x: step(x, threshold),x)), label="step")
plt.plot(x, list(map(lambda x: step(x, threshold),x)), label="derivative")
plt.title("Step Function (dataaspirant.com)")
plt.legend()
plt.show()
```
