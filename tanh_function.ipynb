```
import numpy as np
import matplotlib.pyplot as plt

# Data which will go through activations
x = np.linspace(-10,10,100)
```

## Tanh Function Formula

$f\left(x_{i} \right) = \tanh\left(x_{i}\right),
f^{\prime}\left(x_{i} \right) = 1 - \tanh\left(x_{i}\right)^{2}$

```
def tanh(x, Derivative=False):
  if not Derivative:
    return (np.exp(x)-np.exp(-x)) / (np.exp(x) + np.exp(-x))
  else:
    return 1 - ((np.exp(x)-np.exp(-x)) / (np.exp(x) + np.exp(-x)))**2
```

```
plt.figure(figsize=(12,8))
plt.plot(x, list(map(lambda x: tanh(x),x)), label="hyperbolic")
plt.plot(x, list(map(lambda x: tanh(x, True),x)), label="derivative")
plt.title("Tanh Function Graph (dataaspirant.com)")
plt.legend()
plt.show()
```
